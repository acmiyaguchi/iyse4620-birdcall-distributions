```python
from IPython.display import display
import matplotlib.pyplot as plt
from birdcall_distribution.plot import plot_geometry, plot_grid
from birdcall_distribution.geo import (
    get_americas_geometry,
    AMERICAS_EXTENT,
    generate_grid,
)

geometry = get_americas_geometry()
display(geometry)
plot_geometry(geometry, AMERICAS_EXTENT)
plt.show()
```

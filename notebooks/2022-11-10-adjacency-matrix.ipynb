```python
from birdcall_distribution.geo import (
    get_shape_us_state,
    generate_grid,
    generate_grid_adjaceny_list,
)
import matplotlib.pyplot as plt

ca_shape = get_shape_us_state("California")
grid = generate_grid(ca_shape, (-124, -114, 32, 42), (0.25, 0.25))
adj = generate_grid_adjaceny_list(grid)
```

```python
plt.hist([len(x) for x in adj.values()])
plt.title("distribution of number of neighbors")
plt.show()
```

```python
# write a bit of logic to generate a dense adjacency matrix for the grid
import numpy as np

# convert the keys to integers
keys = sorted(grid.keys())
mapping = {key: i for i, key in enumerate(keys)}

W = np.zeros((len(grid), len(grid)))
for key in keys:
    for neighbor in adj[key]:
        W[mapping[key], mapping[neighbor]] = 1

plt.imshow(W[:25, :25])
plt.show()
plt.imshow(W)
plt.show()
```
